```python
import pandas as pd
```

```python
sales_2012 = pd.read_csv("sales_2012.csv")
sales_2012.head()
```

```python
sales_2013 = pd.read_csv("sales_2013.csv")
sales_2013.head()
```

```python
sales_2014 = pd.read_csv("sales_2014.csv")
sales_2014.head()
```

```python
sales_2015 = pd.read_csv("sales_2015.csv")
sales_2015.head()
```

```python
customer_addresses = pd.read_csv("customer_addresses.csv")
customer_addresses.head()
```

```python
#Union
sales=pd.concat([sales_2012,sales_2013,sales_2014,sales_2015])
sales.head()
```

```python
#Left join
sales_customer_details=sales.merge(customer_addresses, on="Customer ID", how="left")
sales_customer_details.head()
```

```python
#Data type error fixes
sales_customer_details["Revenue"]=sales_customer_details["Revenue"].str.strip("$").astype("float64")
sales_customer_details["Price"]=sales_customer_details["Price"].str.strip("$").astype("float64")
sales_customer_details["Transaction Date"]=pd.to_datetime(sales_customer_details["Transaction Date"])
sales_customer_details.head()
```

```python
#Date range error fix
sales_customer_details.loc[4158,"Transaction Date"]='2015-11-01 00:00:00'
sales_customer_details.head()
```

```python
#Null values fix
sales_customer_details["Product"].fillna("Lomina", inplace=True)
sales_customer_details.head()
```

```python
#Duplicate values fixes
sales_customer_details.drop_duplicates(inplace=True)
column_names = ["Transaction ID","Transaction Date",
                "Product","Company Name","Street Address",
                "City","State","Postal Code"]
sales_customer_details = sales_customer_details.groupby(column_names,as_index=False).sum()
sales_customer_details.reset_index()
```

```python
#Categorical value error fixes
replacements={"Lominade":"Lomina","Tridestand":"Tridesta",
              "Wediicare":"Wedicare","Samatan":"Samtan"}
sales_customer_details=sales_customer_details.replace(replacements)
products=["Lomina","Samtan",
          "Tridesta","Wedicare"]
sales_customer_details=sales_customer_details[sales_customer_details["Product"].isin(products)]
```

```python
#Inconsistent data error fixes
price_hkd=sales_customer_details.loc[sales_customer_details["Transaction Date"].between("2014-01-01",
                                                                                        "2014-12-31"),"Price"]
revenue_hkd=sales_customer_details.loc[sales_customer_details["Transaction Date"].between("2014-01-01",
                                                                                          "2014-12-31"),"Revenue"]
price_usd=round(price_hkd/7.75,2)
revenue_usd=round(price_hkd/7.75,2)
sales_customer_details.loc[sales_customer_details["Transaction Date"].between("2014-01-01",
                                                                              "2014-12-31"),"Price"]=price_usd
sales_customer_details.loc[sales_customer_details["Transaction Date"].between("2014-01-01",
                                                                              "2014-12-31"),"Revenue"]=revenue_hkd
```

```python
cross_column_error =(
    (
    sales_customer_details["Revenue"]-
    (round(sales_customer_details["Price"]*sales_customer_details["Quantity"],2))
    )
    !=0
)
sales_customer_details[cross_column_error]
```
