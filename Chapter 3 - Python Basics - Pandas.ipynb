# Storing Data in Pandas Dataframes

```python
p001 = ["Mouse",24.99,284,"Tolosa","France"]
p002 = ["Keyboard",14.99,128,"Tolosa","France"]
p003 = ["Monitor",199.99,94,"Pro-Tech","Taiwan"]
p004 = ["Webcam",39.99,136,"Brightstone","Japan"]


import numpy as np
p001 = np.array(p001)
p002 = np.array(p002)
p003 = np.array(p003)
p004 = np.array(p004)

products_np = np.array([p001,p002,p003,p004])
print(products_np)
```

# Adding Data to Pandas Dataframes

```python
import pandas as pd

products_dict={"Product":["Mouse","Keyboard","Monitor","Webcam"],
             "Price":[24.99,14.99,199.99,39.99],
             "Stock":[284,128,94,136],
             "Manufacturuer":["Tolosa","Tolosa","Pro-Tech","Brightstone"],
             "Origin":["France","France","Taiwan","Japan"]
        }
product_id=["p001","p002","p003","p004"]

products_pd=pd.DataFrame(products_dict)
products_pd.index=product_id
products_pd
```

```python
products_pd["Color"] = ["Brown","Purple","Gray","Black"]
products_pd
```

```python
products_pd.loc["p005"] = ["Speakers",54.99,14,"Brightstone","Japan","Navy"]
products_pd
```
