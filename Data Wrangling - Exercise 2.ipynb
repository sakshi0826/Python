# 1. Connect to dataset with map coordinates

```python
import pandas as pd
vineyards=pd.read_csv("vineyards.csv")
vineyards
```

```python
from urllib.request import urlretrieve

api_request = "https://ac-101708228-virtuoso-prod.s3.amazonaws.com/uploads/download/119/city_lat_long.csv"

urlretrieve(api_request, "city_lat_long.csv")

city_lat_long = pd.read_csv("city_lat_long.csv")
city_lat_long
```

# 2. Connect to HTML table to get country codes

```python
html_tables = pd.read_html("https://www.iban.com/country-codes")
alpha2_code=html_tables[0]
alpha2_code.drop(["Alpha-3 code","Numeric"],inplace=True,axis=1)
alpha2_code
```

# 3. Combine data streams

```python
alpha2_code["Country"] = alpha2_code["Country"].replace("United States of America (the)","United States")
```

```python
vineyards_alpha2=vineyards.merge(alpha2_code, on="Country", how="left")
vineyards_alpha2
```

```python
vineyards_city_lat_long=vineyards_alpha2.merge(city_lat_long, on=["Alpha-2 code","City"], how="left")
vineyards_city_lat_long.drop("Alpha-2 code",inplace=True,axis=1)
vineyards_city_lat_long
```

# 4. Get minimum temperature forecasts with weather API

```python
from requests import get
core_api ="https://api.weatherapi.com/v1"
api_key ="6306fca418f54e9dad0194450222107"
for row in range(0, len(vineyards_city_lat_long)):
    lat_long = vineyards_city_lat_long.loc[row, "lat,long"]
    url = core_api + "/forecast.json?key=" + api_key + "&q="+ lat_long +"&days=3"
    api_request = get(url)
    request_data = api_request.json()
    vineyard_forecast_day1 = request_data["forecast"]["forecastday"][0]["day"]["mintemp_c"]
    vineyard_forecast_day2 = request_data["forecast"]["forecastday"][1]["day"]["mintemp_c"]
    vineyard_forecast_day3 = request_data["forecast"]["forecastday"][2]["day"]["mintemp_c"]
print(vineyard_forecast_day1)
print(vineyard_forecast_day2)
print(vineyard_forecast_day3)
```
