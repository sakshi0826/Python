# Connecting the files on the Internet

```python
import pandas as pd
sales_dataset = pd.read_csv("sales_dataset.csv")
sales_dataset.head(5)
```

```python
from urllib.request import urlretrieve
url = "https://ac-101708228-virtuoso-prod.s3.amazonaws.com/uploads/download/65/timezones_dataset.csv"
urlretrieve(url, "timezones_dataset.csv")
timezones_dataset = pd.read_csv("timezones_dataset.csv")
```

```python
sales_timezone=sales_dataset.merge(timezones_dataset, on=["City","Country"], how="inner")
sales_timezone.head(5)
```

```python
sales_timezone.drop(["Latitude","Longitude","Population","Elevation","Timezone"], inplace=True, axis=1)
sales_timezone.head(5)
```

```python
sales_timezone["Transaction Date"]=pd.to_datetime(sales_timezone["Transaction Date"])
sales_timezone.head(5)
```

```python
from datetime import timedelta
sales_timezone["GMT_offset"]=pd.to_timedelta(sales_timezone['GMT_offset'],"h")
sales_timezone.head(5)
```

```python
sales_timezone["Transaction Date"]=sales_timezone["Transaction Date"]+sales_timezone["GMT_offset"]
sales_timezone.head(5)
```

```python
sales_timezone.drop("GMT_offset", inplace=True, axis=1)
sales_timezone.head(5)
```

```python
html_tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_circulating_currencies")
html_tables
```

```python
currencies=html_tables[0]
currencies
```

```python
currencies.drop(["Currency[1][2]","Symbol[D] orAbbrev.[3]","Fractionalunit","Numberto basic"], inplace=True, axis=1)
currencies.head()
```

```python
currencies.rename(columns={"State or territory[1]":"Country","ISO code[2]":"Currency Code"}, inplace=True)
currencies.head()
```

```python
sales_timezone["Country"] = sales_timezone["Country"].replace("Democratic Republic of the Congo",
                                                              "Congo, Democratic Republic of the")

sales_timezone["Country"] = sales_timezone["Country"].replace("South Korea",
                                                              "Korea, South")
```

```python
sales_currency=sales_timezone.merge(currencies, on="Country", how="left")
sales_currency[sales_currency["Currency Code"].isna()]
```

```python
from requests import get

sales_currency["Price"]=sales_currency["Price"].astype("float64")

core_api = 'https://api.currencyscoop.com/v1/historical?'
api_key = "d647c8e2d1761955ca81089d6234bc6e"


for row in range(0,len(sales_currency)):
    date=str(sales_currency.loc[row,"Transaction Date"].date())
    currency=sales_currency.loc[row,"Currency Code"]
    url = core_api + "api_key=" + api_key + "&base=" + currency + "&date=" + date
    api_request = get(url)
    request_data = api_request.json()
    conversion_rate=request_data["response"]["rates"]["EUR"]
    sales_currency.loc[row,"Price"]=round(sales_currency.loc[row,"Price"]*conversion_rate,2)
    if row%10==0:
        print(str(row)+"% complete.")

print("100% complete")
sales_currency
```
